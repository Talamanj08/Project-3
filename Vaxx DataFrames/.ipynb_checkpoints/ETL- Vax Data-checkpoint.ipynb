```python
import pandas as pd
from pathlib import Path
```

```python
Vaccinations= Path("Data/COVID-19_Vaccinations_in_the_United_States_County_smaller.csv")
```

```python
Vax_df= pd.read_csv(Vaccinations, encoding="ISO-8859-1")
```

```python
Vax_df.head()
```

```python
columns_to_delete = ['Census2019', 'Census2019_5PlusPop', 'Census2019_5to17Pop', 'Census2019_12PlusPop', 'Census2019_18PlusPop', 'Census2019_65PlusPop','FIPS','MMWR_week', 'Completeness_pct']
Vax_df = Vax_df.drop(columns=columns_to_delete)
```

```python
Vax_df.head()
```

```python
Vax_df = Vax_df.rename(columns={"Recip_County":"Counties", "Recip_State":"State"})
Vax_df.head()
```

```python
Vax_df['Date'] = pd.to_datetime(Vax_df['Date'], format='%m/%d/%Y')
Vax_byState_df = Vax_df.groupby([Vax_df['Date'].dt.year, 'State']).sum()
pd.set_option('display.float_format', '{:.2f}'.format)
Vax_byState_df
```

```python
len(Vax_byState_df)
```

```python
unique_states = Vax_byState_df.index.get_level_values('State').unique()
len(unique_states)
```

```python
level_to_delete = 'State'
values_to_delete = ['AS', 'FM', 'GU', 'MH', 'MP', 'PR', 'PW', 'UNK', 'VI', 'DC', 'HI', 'AK']
Vax_byState_df = Vax_byState_df.drop(values_to_delete, level=level_to_delete)
Vax_byState_df
```

```python
states = Vax_byState_df.index.get_level_values('State').unique()
len(states)
```

```python
len(Vax_byState_df)
```

```python
years = Vax_byState_df.index.get_level_values('Date').unique()
print(years)
```

```python
Vax_byState_2021 = Vax_byState_df.groupby(level='Date').get_group(2021)
Vax_byState_2022 = Vax_byState_df.groupby(level='Date').get_group(2022)
Vax_byState_2023 = Vax_byState_df.groupby(level='Date').get_group(2023)
```

```python
Vax_byState_df.to_csv("Data/Vax_byState_ALLYears.csv", index=False, encoding="ISO-8859-1")
Vax_byState_2021.to_csv("Data/Vax_byState_2021.csv", index=False, encoding="ISO-8859-1")
Vax_byState_2022.to_csv("Data/Vax_byState_2022.csv", index=False, encoding="ISO-8859-1")
Vax_byState_2023.to_csv("Data/Vax_byState_2023.csv", index=False, encoding="ISO-8859-1")
```
